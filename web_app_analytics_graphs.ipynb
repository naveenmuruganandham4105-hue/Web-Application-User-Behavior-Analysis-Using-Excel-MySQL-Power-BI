```
# User-defined function to categorize engagement
def categorize_engagement(time):
    if time < 600:          # less than 10 minutes
        return 'Low'
    elif time < 1800:       # 10–30 minutes
        return 'Medium'
    else:
        return 'High'

# Creating new column
df['engagement'] = df['session_duration_sec'].apply(categorize_engagement)

# View result
df[['session_duration_sec', 'engagement']].head()
```

```
import pandas as pd

# Load dataset
df = pd.read_csv("web_app_analytics_100_records.csv")

# Count of each device
device_count = df['device'].value_counts()

# Print output
print(device_count)
```

```
df.groupby('device')['session_duration_sec'].mean().plot(kind='bar', color='yellow')

plt.title("Avg Session Duration by Device")
plt.xlabel("Device")
plt.ylabel("Session Duration (sec)")
plt.show()
```

```
import matplotlib.pyplot as plt

for d in df['device'].unique():
    subset = df[df['device'] == d]
    plt.hist(subset['session_duration_sec'], alpha=0.5, label=d, color='yellow')

plt.title("Histogram by Device")
plt.xlabel("Session Duration")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```
df.groupby('device')['clicks'].mean().plot(kind='line', marker='o', color='yellow')

plt.title("Avg Clicks by Device")
plt.xlabel("Device")
plt.ylabel("Clicks")
plt.show()
```

```
import matplotlib.pyplot as plt

plt.hist(df['session_duration_sec'], color='yellow')

plt.title("Session Duration Distribution")
plt.xlabel("Seconds")
plt.ylabel("Frequency")
plt.show()
```

```
import matplotlib.pyplot as plt

plt.hist(df['clicks'], bins=5, color='yellow')

plt.title("Distribution of Clicks")
plt.xlabel("Number of Clicks")
plt.ylabel("Frequency")
plt.show()
```

```
import matplotlib.pyplot as plt

plt.scatter(df['session_duration_sec'], df['clicks'], color='yellow')

plt.title("Session Duration vs Clicks")
plt.xlabel("Session Duration (sec)")
plt.ylabel("Clicks")
plt.show()
```

```
import matplotlib.pyplot as plt

plt.plot(df['clicks'], df['scroll_percent'], color='yellow', marker='o')

plt.title("Clicks vs Scroll Percentage")
plt.xlabel("Clicks")
plt.ylabel("Scroll Percentage")
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

correlation = df[['session_duration_sec','clicks','scroll_percent','bounce','conversion']].corr()

sns.heatmap(correlation, annot=True, cmap='YlOrBr')

plt.title("Correlation Heatmap")
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['session_duration_sec','clicks','scroll_percent','bounce','conversion']])

plt.show()
```


1. Users who spend more session duration tend to have higher clicks and scroll percentage, indicating better engagement.
2. Clicks and scroll percentage show a positive relationship, meaning users who interact more also scroll more.
3. Certain pages like Home, Product, and Checkout receive more user activity, indicating user preference.
4. Mobile and Desktop devices contribute more users compared to other devices.
5. Users with higher bounce rate show lower interaction, indicating poor engagement.
6. Higher session duration is associated with higher number of clicks, showing active user behavior.
7. Some users achieve high scroll percentage even with moderate session duration.
8. Conversion rate varies among users, indicating different levels of user interest.
9. User interaction metrics like clicks and scroll percentage show variability across users.
10. Session duration, clicks, and scroll percentage together influence conversion behavior.



**description**

The dataset web_app_analytics_100_records contains information about user interactions within a web application. It includes various metrics that help analyze user behavior, engagement, and performance of the application.

Each record represents a user session and provides details such as the page visited, type of device used, session duration, number of clicks, scroll percentage, bounce rate, and conversion status.

The dataset is useful for performing univariate, bivariate, and multivariate analysis to understand how users interact with the application. It helps in identifying patterns such as user engagement levels, popular pages, device usage, and factors influencing conversions.

By analyzing this dataset, we can gain insights into user behavior, improve user experience, and optimize the performance of the web application.

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from Github and many other sources. To learn more about importing data, and how Colab can be used for data science, see the links below under [Working with Data](#working-with-data).

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
